```python
import pandas as pd
tta_path = './tta_test/tta_double/'
pd.set_option('display.float_format', '{:.2f}'.format) # 항상 float 형식으로 출력
```

```python
answer_list = [2406, 1652, 1596, 1464, 1222, 640, 478, 334, 326, 288, 240, 162, 492, 300, 322, 290, 254, 134]
```

```python
abs_df = pd.DataFrame()
abs_df['abs_sum']=[]
```

## ('HueSaturationValue', 'horizontal')

```python
df1 = pd.read_csv(tta_path+"hsv+hor1.csv")
df2 = pd.read_csv(tta_path+"hsv+hor2.csv")
df3 = pd.read_csv(tta_path+"hsv+hor3.csv")
df4 = pd.read_csv(tta_path+"hsv+hor4.csv")
df5 = pd.read_csv(tta_path+"hsv+hor5.csv")
```

```python
hsv_hor = pd.DataFrame(answer_list, columns=['answer'])
hsv_hor['result1'] = df1.ans.value_counts()
hsv_hor['result2'] = df2.ans.value_counts()
hsv_hor['result3'] = df3.ans.value_counts()
hsv_hor['result4'] = df4.ans.value_counts()
hsv_hor['result5'] = df5.ans.value_counts()
```

```python
hsv_hor['average'] = (hsv_hor['result1']+hsv_hor['result2']+hsv_hor['result3']+hsv_hor['result4']+hsv_hor['result5'])/5
hsv_hor['final'] = (hsv_hor['answer'])\
                        - (hsv_hor['result1']+hsv_hor['result2']+hsv_hor['result3']+hsv_hor['result4']+hsv_hor['result5'])/5
```

```python
hsv_hor
```

```python
print(hsv_hor[["answer", "average","final"]].describe())
print(f"\nsum of final's abs: {sum(abs(hsv_hor['final']))}")
abs_df.loc['hsv_hor']=sum(abs(hsv_hor['final']))
```
