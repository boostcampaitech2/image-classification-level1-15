```python
import pandas as pd
tta_path = './tta_test/'
pd.set_option('display.float_format', '{:.2f}'.format) # 항상 float 형식으로 출력
```

```python
answer_list = [2406, 1652, 1596, 1464, 1222, 640, 478, 334, 326, 288, 240, 162, 492, 300, 322, 290, 254, 134]
```

```python
abs_df = pd.DataFrame()
abs_df['abs_sum']=[]
```

```python
abs_df.loc['orig']=1660.0
abs_df.loc['horizontal']=1615.2
abs_df.loc['ShiftScaleRotate']=1663.6
abs_df.loc['gaussian']=1776.4
abs_df.loc['MutiplicativeNoise']=2256.4
abs_df.loc['iso']=2003.6
```

## coarsedropout

```python
df1 = pd.read_csv(tta_path+"cdo1.csv")
df2 = pd.read_csv(tta_path+"cdo2.csv")
df3 = pd.read_csv(tta_path+"cdo3.csv")
df4 = pd.read_csv(tta_path+"cdo4.csv")
df5 = pd.read_csv(tta_path+"cdo5.csv")
```

```python
coarsedropout = pd.DataFrame(answer_list, columns=['answer'])
coarsedropout['result1'] = df1.ans.value_counts()
coarsedropout['result2'] = df2.ans.value_counts()
coarsedropout['result3'] = df3.ans.value_counts()
coarsedropout['result4'] = df4.ans.value_counts()
coarsedropout['result5'] = df5.ans.value_counts()
```

```python
coarsedropout['average'] = (coarsedropout['result1']+coarsedropout['result2']+coarsedropout['result3']+coarsedropout['result4']+coarsedropout['result5'])/5
coarsedropout['final'] = (coarsedropout['answer'])\
                        - (coarsedropout['result1']+coarsedropout['result2']+coarsedropout['result3']+coarsedropout['result4']+coarsedropout['result5'])/5
```

```python
coarsedropout
```

```python
print(coarsedropout[["answer", "average","final"]].describe())
print(f"\nsum of final's abs: {sum(abs(coarsedropout['final']))}")
abs_df.loc['coarsedropout']=sum(abs(coarsedropout['final']))
```
